```python
import matplotlib.pyplot as plt
import matplotlib.patheffects as path
import numpy as np

n = 3000
fig = plt.figure(figsize=(10, 10))
ax = fig.add_subplot(111, projection='3d')
ax.set_facecolor('Black')

xs, ys, zs = np.zeros(n), np.zeros(n), np.zeros(n)
for i in range(n):
    t = np.sqrt(1 / n * i)
    z = 1 - t
    x = t*np.cos(t*1000)
    y = t*np.sin(t*1000)
    xs[i], ys[i], zs[i] = x, y, z
color = np.array([np.random.random_sample() for i in range(n)])
ax.scatter(xs, ys, zs, c=color, marker='o', s=15, cmap='Greens')
ax.scatter([0], [0], [1.001], marker='*', s=500, color='Yellow')
txt = ax.text(0.03,0.03,0.3,'Merry\nChristmas', color='White', size = 15, ha='center', weight = 'bold')
txt.set_path_effects([path.withStroke(linewidth=3, foreground='black')])
ax.axis([-2,2,-2,2])

plt.show()
```
